# Script to check ROI Masks in Ground Truth

```python
import os, sys
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns

from tqdm import tqdm
from PIL import Image
```

```python
cols = list(pd.read_csv("mass_case_description_train_set.csv", nrows =1))
print(cols)
df= pd.read_csv("mass_case_description_train_set.csv", usecols =['pathology','cropped image file path','ROI mask file path'])
print(df)
```

### Images of small size are cropped images it seems there is some error in csv file provided by CBIS DDSM

```python
cont = 0
for i in range(0, len(df)):
    mask = df.iloc[i]['ROI mask file path'].split('\n')
    mask_path = os.path.join('./data/train', mask[0])
    crop = df.iloc[i]['cropped image file path'].split('\n')
    crop_path = os.path.join('./data/train', crop[0])
    size_mask = os.path.getsize(mask_path)
    size_crop = os.path.getsize(crop_path)
    print(size_mask, size_crop)
    if size_crop > size_mask:
        cont+=1
        print("incorrect: ", mask)
        
print('Total errors:', cont)

            
```
